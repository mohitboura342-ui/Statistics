## _Key Topics_
- ***1.Sampling Distribution:*** This is the probally distribution of a given statistic based on a random sample. It's crucial for understanding how the statistic might vary and for estimating the uncertainty of an inference.
- ***2.Point Estimation:*** This invloves using sample data to calculate a single value (known as a point estimate) which serves as a "best guess" of an unknown population parameter (like the population mean or proportion).

- ***3.Interval Estimation:***  Instead of estimating a parameter by a single value, interval estimation provides a range of values [confidence interval], written which the parameter is expected to lie, with a certain level of confidence (e.g, 95%).
- ***4.Hypothesis Testing:*** This is a formal procedure used to test whether a hypothesis about a parameter is true. It starts with a null hypothesis (H0) representing a default position or status quo, and an alternative hypothesis (H1) that is considered if the data provide sufficient evidence against H0.

## ***Sampling Distribution of a Statistics***
- Statistic: A metric calculated for a sample of data drawn from a larger population. Eg - Sample mean
- Sampling distribution of a statistic refers to the distribution of some sample statistic over many samples drawn from the same population
- A sample is drawn with the goal of measuring something (with a sample statistic). Since our estimate or model is based on a sample, it might be in error; it might be different if we were to draw a different sample.
- We are therefore interested in how different it might be - a key concern is sampling variability. If we had lots of data, we could draw additional
samples and observe the distribution of a sample statistic directly.

## ***Central Limit Theorem***
-  The Central Limit Theorem states that the means drawn from multiple samples will resemble the Normal distribution even if the population is not normally distributed, provided that the sample size is large enough and the departure of the data from normality is not too great

## ***Standard Error***
- The standard error is a single metric that sums up the variabilty in the sampling distribution for a statistic.
- $\text{SE}_{\bar{x}} = \frac{s}{\sqrt{n}}$
- As the sample size increases, the standard error decreases.

## ***Estimator $ Estimates***
- An estimator is a rule or a mathematical formula used to calculate an estimate of a population parameter based on sample data
- For example, the sample mean & is an estimator of the population mean u. The formula for the sample mean is:
- $$\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i$$

## <b>_Types of Estimates_
- ***Point Estimate:*** A point estimate is a single value that serves as an approximation of a population parameter. Common point estimates include the sample mean, median, proportion, and standard deviation, which estimate the corresponding population parameters.
- ***Confidence Interval:*** A confidence interval, on the other hand, provides a range of values within which the true population parameter is likely to fall, with a certain level of confidence.
- ***Eg-*** Continuing with the height estimation, suppose the average beight of your sample is 175 cm, and you calculate a 95% confidence interval from 170 cm to 180 cm. This interval suggests that you are 95% confident that the true average height of all adult males in the city is between 170 cm and 180 cm.
- We build the confidence interval around the point estimate.
- A confidence interval is constructed on the sampling distribution of a statistic.
- (1-𝛼) is the level of confidence. We are (1-a)1*00% confident that the population parameter will fall in the specified interval. Common alphos are: 0.01, 0.05, 0.1.

### Confidence Intervals
- General formula to comute confidence intervals:
  <b>[Point estimate - Margin of error, Point estimate + Margin of error]
- The square brackets [] imply that the starting and ending points are included in the interval.
- ***Margin of error:*** The margin of error is a statistic expressing the amount of random sampling error in the results of a survey.
- The margin of error can be calculated either using the Z - distribution (standard normal distribution) on the Student's T distribution.
- $z_{\alpha/2} \frac{\sigma}{\sqrt{n}}$
- $t_{\alpha/2} \frac{s}{\sqrt{n}}$ 

### *Example*
- Let's say we want to estimate the average height of adult men in a city. We take a random sar 100 men and find their average height to be 175 cm with a standard deviation of 10 cm. We calculate a 95% confidence interval for the true average height.

1. ***Calculate the standard error (SE):***

  $$SE_{\bar{x}} = \frac{\sigma}{\sqrt{n}}$$
  
  - $\sigma$ is the population ***standard deviation.***
  - $n$ is the sample size ***(number of observations).***

   - SE = $\frac{10}{\sqrt{100}}$ = 1cm


2. ***Determine the critical value for a 95% confidence level:*** For a 95% confidence level, the
value (Z-score) is approximately 1.96
                   
3. ***Calculate the margin of error (ME):***

    - ***ME = critical value × SE*** = 1.96 × 1 = 1.96cm

4. ***Compute the confidence interval:***

   - ***Confidence Interval = sample mean ± ME = 175 ± 1.96***

   - ***Confidence Interval = [173.04 cm, 176.96 cm]***

### *Interpretation*

- We are 95% confident that the true average height of adult men in the city is between 173.04 cm and 176.96 cm.

### *Conclusion*

- A confidence interval provides a range within which we believe the true population parameter lies,based on our sample data and the specified confidence level. This range helps in making inferences about the population parameter with a quantifiable level of uncertainty. 

### *Confidence Interval*

- ***The higher the level of confidence, the wider the interval. Also, the smaller the sample, the wider the interval*** (i.e., the more uncertainty). Both make sense: the more confident you want to be, and the less data you have, the wider you must make the confidence interval to be sufficiently assured of capturing the true value

## ***Approaches for Conducting a Hypothesis Test***
- There are mainly two approaches for conducting a hypothesis test.
- ***Critical value approach:*** Compares the test statistics to a per-defined critical value based on the chosen significanace level($\mathbf{\alpha}$) and degrees of freedom. If the test statistics falls outside the critical region the null hypothesis is rejected.
- ***P-value approach:*** Calculates the probability of observing a test statistics as extreme as the one calculated,  assuming the null hypothesis is true. A low p-value < $\mathbf{\alpha}$ provides evidence against the null hypothesis.

### ***Steps for Critical Value Approach***
- State your null (H0) alternative (H1) hypothesis. Clearly define what you are testing for.
- Choose your significance level ($\mathbf{\alpha}$). This is the acceptable probability of falsely rejecting the null hypothesis (usually set at 0,05).
- Calculate the test statistic. This depends on your specific test (e.g, z-score for z-test,. chi-square statistic for chi-square test). Use your sample data and relevant calculations.
- Determine the critical value(s). Refer to a statistical table or software based on your test, degrees of freedom (ussally samole size-1) and chosen $\mathbf{\alpha}$. This defines the rejection region(s) where the null hypothesis is unlikely to be true.
- Compare the test statistic to the critical value(s). If the test statistic falls outside the rejection region(s) (e.g, greter than the upper critical value for a one-tailed test), reject the null hypothesis. If it falls within the rejection region(s), fail to reject the null hypothesis.

## ***Steps for p-value approach***
- State your null (H0) and alternative (H1) hypothesis. Same as in the critical value approach.
- Calculate the test statistic. Same as in the critical value approach.
- Calculate the p-value. This is the probability of observing a test statistic as extreme  as the one calculated, assuming the null hypothesis is true.
- Compare the p-value to your chosen $\mathbf{\alpha}$. If the p-value is less than $\mathbf{\alpha}$, reject the null hypthesis. If it's greater than $\mathbf{\alpha}$, fail to reject the null hypothesis.

## ***P-value***
- P-value is the probability that, given a chance model, results as extreme as the observed results could occur.
- Let's understand this through an example.
- ***Null hypothesis (H0):*** The coin is fair. meaning it has a 50% chance of landing heads and 50% chance of landing tails.
- ***Alternate Hypothesis(H1):*** The coin is unfair
- You flip the coin 10 times and observe 9 heads. This seems unusual if the coin is truly unfair.
- P-value: The p-value calculates the probability of getting 9 or more heads (or even more extreme results) in 10 flips, assuming the coin is fair (H0 is true).
- If the p-value is very low (say 0.01), it means there's only 1% chance of seeing such a result purely by chance if the coin is fair. This suggests that H0 might be wrong, and the coin might actually be biased towards heads.
- Typically a, p-value less than the chosen significance level is considered level is considered significance evidance against the null hypothesis.

- Common misconeption about p-value:
- *The probability that the resutls is due to chance*

- This is what actually a p-value represents:
- The probabilty that, given a chance model, results as extreme as the observed results could occur.
- A low p-value simply suggests the observed datat is unlikely under the assumption of the null hypothesis being true, but it dosen't guarantee the truth of the alternative.

## ***Chi-Square Test***
- The chi-square test is a statistical test commonly used to evaluate whether there is a significant association between two categorical variables or to test the goodness-of-fit between as observed distribution and an expected theoretical distribution.
- There are two main types of Chi-Square tests: The Chi-Square test of independence and the Chi-Square goodness-of-fit test.

- Chi-Square test is a non-parametric test used on nominal data
- It aims to determine whether significant difference exists among groups of data.
- The chi-square test can only be used if:
- The experminta data or sample observations are independent of each other.
- The data collected must be drawn at random from the universe or population.
- These data must be presented in original units, not in ratio or percentage form.
- There must be at least 50 observations in the sample. Also, the number of observations within each cell should not be less than 5.

## ***Using the Chi-Square Test***
- $$\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}$$
- $\chi^2$ is the Chi-Square test statistic.
- $\sum$ (Sigma) is the summation operator, meaning you sum the result for every category or cell in your data.
- $O_i$ is the Observed Frequency (the actual count of observations in each category from your sample data).
- $E_i$ is the Expected Frequency (the count you would expect to see in each category if the null hypothesis were true, i.e., if there were no relationship or difference).
- Steps:
  1. Compute the expected frequencies. 
  2. Compute the chi-squared test statistic.
  3. Find the degrees of the freedom as:
     - df = k - 1, where k is the no of categories
  4. Find the critical value from the chi-square table
  5. If the test statistic is greater than the critical value, then H0 is rejected.

## ***Correlation***

- ***What is Correlation?***
- Correlation is a statistical measure that describes the strength and direction of a relationship between two variables. It quantifies how changes in one variable are associated with changes in another. The correlation coefficient, usually denoted as r, ***ranges from -1 to +1:***
- ***r = 1:*** Perfect positive correlation (as one variable increases, the other increases proportionally).
- ***r = -1:*** Perfect negative correlation (as one variable increases, the other decreases proportionally).
- ***r = 0:*** No correlation (no linear relationship between the variables).

## ***Correlation Formula***

- ***Correlation Formula***
- The most common measure of correlation is the Pearson correlation coefficient, which quantifies the linear
relationship between two variables. The formula for Pearson's correlation coefficient (r) is:

$$\huge r = \frac{n (\sum xy) - (\sum x) (\sum y)}{\sqrt{[n \sum x^2 - (\sum x)^2][n \sum y^2 - (\sum y)^2]}}$$

- Key Terms in the Formula
  - $r$: The Pearson correlation coefficient. Its value is always between $-1$ and $+1$.
  - $n$: The number of data points (pairs of $x$ and $y$ values).
  - $\sum x$: The sum of the $x$ values.
  - $\sum y$: The sum of the $y$ values.
  - $\sum xy$: The sum of the products of $x$ and $y$ for each data point.
  - $\sum x^2$: The sum of the squared $x$ values.
  - $\sum y^2$: The sum of the squared $y$ values.
  - $(\sum x)^2$: The square of the total sum of the $x$ values (calculated by squaring $\sum x$).
  - $(\sum y)^2$: The square of the total sum of the $y$ values (calculated by squaring $\sum y$).

## ***Why is Correlation Used?***
- Correlation is used in various fields for several reasons:
1. ***Identifing Relationships:*** To determine if and how strongly two variables are related.
2. ***Predictive Analysis:*** To predict the value of one variable based on the value of another.
3. ***Data Simpilification:*** To predict the complexity of the data by identifying relationship and dependancies.
4. ***Feature Selection:*** In machine learning, to select relevant features by identifying amd removing highly correlated variables.

## ***Correlation and Causation***

- The phrase ***"correlation is not causation"*** means that just because two variables are correlated [i.e., they move together in some way), it does not necessarily mean that one variable causes the other. Correlation indicates a relationship or association between variables, but causation indicates that one variable directly affects the other. Here are a few examples to illustrate this concept:

  
- ***Example 1: Ice Cream Sales and Drowning Incidents***
  - ***Correlation:*** Studies may show that ice cream sales and drowning incidents both increase during the summer months
  - ***Expianation:*** While there is a correlation between ice cream sales and drowning incidents, buying ice cream does not cause drowning. The actual cause is the hot weather, which leads to more people buying ice cream and more people swimming, thus increasing the risk of drowning

  
- ***Example 2: Shoe Size and Reading Ability in Children***
  - ***Correlation:*** Data might show a positive correlation between shoe size and reading ability in young children
  - ***Explanation:*** Larger shoe sizes do not cause better reading ability. The underlying factor is age; older children tend to have larger shoe sizes and better reading skills because they have had more time to develop these skills.
