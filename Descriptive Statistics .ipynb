## What is Data?

Data refers to information or observations that are collected, recorded, or analyzed to draw conclusions, make interfences, or support decision-making.

### _Types of Data:_

***1.Categorical Data (Classified Data)***
- Categorical data represents groups or categories.
- ***Examples-***
1. Car brands : Audi, BMW, and Mercedes.
2. Answers to yes/no questions : yes and no


***2.Numerical Data***
- Numerical data represents numbers.
- ***Examples-***
1. Marks of a students in English.
2. Sales of a company.
3. Income of employees of a company.
4. Height of each student in a class.

### Numerical Data

***1.Discrete data***
- Discrete data take on specific, distinct values and are counted in individual units. The values are often integers, and are seperate from each other.
- ***Eg.:***
1. Number of children in a school.
2. Number of books in your libraray.
3. Number of cases a Lawayer has won.

***2. Continuous data***
- Continuous data can take on any value within a given range.
- ***Eg:***
1. Height of an individual 175.78 , 150.05cm
2. Monthly Expenditure of a company. Rs.120000034.34 , Rs.23432.34

### Levels of Measurement 

- The level of measurement refers to how precisely data has been measured.

- This data can be measured in various ways:
  1. a.10-19k  b.20-29k  c.30-39k  d.50k
  2. a.Heigh   b.Medium  c.Low
  3. You can ask the exact figrues

***Nominal data***
- The nominal level represents categories that cannot be put in any order.
- Eg:
  1. Car brands : Maruti, Honda , Mahindra, Hyundai
  2. Subjects : Maths, Humanities , Commerce , Biology


***Ordinal data***
- Eg:
  1. Rating of a food - Neutral , Tasty
  2. Programing level - Beginner , Intermediate , Advanced

***Interval data***
- This level of measurement has no true zero. This means that a zero does not mean absence of a property.
- Eg:
  1. Temperature in Celsius or Fahrenheit : 0c or 0F does not mean the lowest temperature.
  2. Credit scores range from 200-800. Here 0 does not mean true zero.


***Ratio data***
- This level of measurement has a true zero.
- Eg:
  1. Temperature in Kelvin: 0k means the lowest possible temperature
  2. Height of an object - if the height of an object is 0cm then it means no height

<font size = '5'>***Measures of Central Tendancy***

- Central tendancy is a way to describe the "center" of "typical" value of a set of numbers. It gives you a sense of where most of the data is concntrated.
- You can "make sense" of the data without looking at all the values or observations by just looking at the central tendency.
- A measure of central tendancy is a summary measure of your column.

- There have been several measures of central tendency developed in statistical theory. The most commonly used ones include:
  - Mean or Arithmetic mean
  - Median
  - Mode

<font size = '5'>***Mean***
- It can be calculated as -
- $\bar{x} = \frac{\sum x_i}{n}$

<font size = '5'>***Median***

- It can be calculated by arranging values in either ascending or descending order and then finding the middle value.
- For odd number of values:
  - $\text{Median Position} = \frac{(n+1)}{2} \text{th term}$
- For even numbers of values:
  - $Median = \frac{\left(\frac{n}{2}\right)^{\text{th}} \text{ term} + \left(\frac{n}{2} + 1\right)^{\text{th}} \text{ term}}{2}$

<font size = '5'>***Mode***

- $$\text{Mode} = \text{The value with the highest frequency}$$
- The mode of the data is the most frequently occuring value.
- The mode can be used for nominal data (Categorical Data).

<font size = '5'>***Uses of Measures of Dispersion***

<font size = '5'><b>1.Range
- Range is the difference between the largest and smallest data points in the set.
- Range = Max value - Min value
- For our example -
  - Player 1 = 5 , 15 , 15 , 25
  - Player 2 = 10 , 15 , 15 , 20
- Mean of Player 1 = 15
- Mean of Player 2 = 15
- Range of Player 1 = 25 - 5 = 20
- Range of Player 2 = 20 - 10 = 10
Thus, Player 2 is more consistent than player 1 and has more reliable average.

### ***Problems with the Range***
- ***Sensitive to Outlirs*** : The range is highly sensitive to extreme values (outliers) because it depends only on the maximum and minimum values. A single extreme value can significantly disort the measure.
  <b>Example :<b> Consider the following dataset: [10, 12, 15, 18, 20, 100]. The range is 100 - 10 = 90. However, the majority of values are within a small range , and the large outlier [100] greatly influences the result.

<font size = '5'>***Quartile***
- The first quarutile, represented as Q1, refers to the 25 percentile of the data.
- The second quartile , represented as Q2, refers to the 50 percentile of the data. It is actually the median of the data.
- The third quartile, represented as Q3, refers to the 75 percentile of the data.

<font size = '5'>***Interquartile Range(IQR)***
- Interquartile Range is the difference between the first (Q1) and the third quartile (Q3).
- It describes the middle 50% of the observations. If the middle 50% of the data points are spaced wide apart, their interquartile range will be large.
- ***IQR = Q3 - Q1***
- <b>The IQR is primarily  used for detecting outliers. Values below ***Q1 - 1.5IQR*** and above ***Q3 + 1.5IQR*** are considered as outliers.

<font size = '5'>***3.Variance***
- Variance measures the spread of the data around the mean.
- It is calculated as the average of the squared differences betweeen each data point and the mean of the dataset.

- If data values are very different → high variance

 - If data values are similar → low variance

Where:
- $X_i$ is each individual data point.
- $\mu$  is the population mean (average).
- $N$ is the number of data points (size of the population).
- $\bar{x}$ is the sample mean (average).
- $n$ is the sample size (size of the sample).
- $\text{s}^2$ is the sample variance.
- $\sigma^2$ is the population variance.

### ***Drawback of the variance***
- Varince is in squared units of the original data which makes it harder to relate with the original data.

<font size = '5'>***4.Standard Deviation***
- Standard deviation is the most popular masure of dispersion.
- It is simply the square root of the variance.
- It overcomes the major drawback of the variance as it is expressed in the unit of the original data.
- 

# ✅ Difference Between Variance and Standard Deviation
 - ***variance***

Measures how far data points spread from the mean.

Uses squared differences.

Unit is squared (e.g., cm², kg²).

Harder to interpret.

- ***Standard Deviation (SD)***

Square root of variance.

Measures spread in the same unit as the data.

Easier to understand and compare.
Variance shows spread in squared units, while Standard Deviation shows spread in original units (√variance).

## Z-score and Emperical Loop

###### Z-score
- Also called the *standard score*.
- It gives an idea of how far from the mean a data point is.
- Z-score is the number of standard deviations an observed value is above or below the mean value.
- When using z-score it is assumed that the data is normally distributed.
- 





- 🌟 Why Z-Score is Useful?
1️⃣ Detecting Outliers

If z-score > 3 or < –3 → usually considered an outlier

2️⃣ Standardization

Converts different scales into a common scale

Useful in machine learning (e.g., KNN, Linear Regression)

3️⃣ Comparison

Helps compare values from different distributions

- Z-score batata hai ki koi data point mean se kitne standard deviations door hai.

***Empirical Rule***
- The empirical rule for for normal distribution or the 68-95-99.7. Rule states 68%, 95%, and 99.7% of the values lie within one, two, and three standard deviations of the mean, respectively.

***Coefficient of Variation***
- The coefficient of variation is a measure of dispersion used to compare the dispersion between data on different scales.
- When data is in different scales, it's not suitable to use the standard deviation to compare the dispersion of the two datasets.
- The coefficient of variation is only calculated for data in ratio scale.

***Kurtosis***
- Kurtosis is a measure of the "tailedness" of a distribution.
- Data sets with high kurtosis tend to have heavy tails, or outliers. Data sets with low kurtosis tend to have lights tails, or lack of outliers.
- Kurtosis ranges from 1 to infinity.
- A normal distribution has a kurtosis of 3.
- Kurtosis is sometimes seem as the "peakedness" of a distribution which is incorrect.

- Here both the three data sets have the same mean and standard deviation and neither are skewed. To compare such distributions we use kurtosis.
- 

- A normal distribution is called ***mesokutic distribution*** and has a kutosis of 3.

### <b> _Leptokurtic and Platykurtic Distributions_
- Distribution with kurtosis greater than 3 are called leptokurtic.
- Distribution with kurtosis less than 3 are called platykurtic.
- 

### _Excess Kurtosis_
- Excess Kurtosis = Kurtosis - 3
- Excess Kurtosis is often used to compare the kurtosis of various distributions to normal distribution.
- It ranges from -2 to infinity.
- A normal distribution has a excess kurtosis = 3 - 3 = 0
- Positive excess Kurtosis is Leptokurtic.
- Negative excess kurtosis is platykurtic.
